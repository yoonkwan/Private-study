```python
### Correlation 알아볼 함수 만들기

from matplotlib.colors import LogNorm

def plot_cor(x,y):
    plt.figure()
    plt.hist2d(x,y, bins=200, cmap='jet', norm=LogNorm())
    plt.colorbar()
    plt.show()
```

```python
Time=[100,200,300,400,500]
z=[]
Metal=[]

for i in Time:
    g100=h.f_rdgal(i,100)
    z100=1/g100['Aexp']-1
    z.append(z100)
    p100=h.f_rdptcl(i,100)[0]
    Metal.append(np.mean(p100['metal']))
```

```python
plt.figure()
plt.scatter(z,Metal,c='b')
plt.xlabel('z')
plt.ylabel('Metallicity')
plt.gca().invert_xaxis()
plt.show()

# 대체로 시간이 흐르면서 metallicity가 증가하는데 z=2.0과 1.0 사이에 감소했었다.
# 외부로의 유출이 있었나...?
```

```python
Time=[100,200,300,400,500]
z=[]
Mass=[]

for i in Time:
    g100=h.f_rdgal(i,100)
    z100=1/g100['Aexp']-1
    z.append(z100)
    Mass.append(g100['Mass_tot'])
```

```python
plt.figure()
plt.scatter(z,Mass,c='b')
plt.xlabel('z')
plt.ylabel('Galaxy Mass')
plt.gca().invert_xaxis()
plt.show()
```

```python
gAll=h.f_rdgal(100,-1)
HaloAll=h.f_rdgal(100, -1, horg='h')
```

```python
from matplotlib import patches

H_Xc=np.array(HaloAll['Xc'], dtype=float)
H_Yc=np.array(HaloAll['Yc'], dtype=float)
H_R=np.array(HaloAll['R_200crit'], dtype=float)

G_Xc=np.array(gAll['Xc'], dtype=float)
G_Yc=np.array(gAll['Yc'], dtype=float)
G_R=np.array(gAll['R_size'], dtype=float)
```

```python
plt.figure(figsize=(15,10))

for i in range(len(H_Xc)):
    plt.scatter(HaloAll['Xc'],HaloAll['Yc'], c='gray', s=HaloAll['R_200crit'][i])

for i in range(len(gAll)):
    plt.scatter(gAll['Xc'],gAll['Yc'], c='b', s=gAll['R_size'][i], alpha=0.7)
    
plt.show()
```

```python
plt.figure(figsize=(15,10))

for i in range(len(H_Xc)):
    c=patches.Circle((H_Xc[i],H_Yc[i]), H_R[i], fc='gray', ec='k')
    plt.gca().add_patch(c)
    
for i in range(len(gAll)):
    plt.scatter(gAll['Xc'],gAll['Yc'], c='b', s=gAll['R_size'][i], alpha=0.7)
    
plt.show()

### 이런 방식으로도 해봤는데, 딱히 좋은 플롯이 그려지지는 않았다...

### 원래는 단순하게 매칭하는게 아니라, 논문을 참고해서 헤일로와 은하를 나타내고,
### 헤일로의 R200 내에 은하가 위치한 것을 확인해서 galaxy group을 찾아보려는 목표였지만, 잘 안된것같다.
### 하나는 원 그리는 방식으로 표현하고, 하나는 scatter로 하다보니 더 안맞는 듯 하다.
```

```python
plt.figure(figsize=(15,10))

for i in range(len(H_Xc)):
    c=patches.Circle((H_Xc[i],H_Yc[i]), H_R[i], fc='gray', ec='k')
    plt.gca().add_patch(c)
    
for i in range(len(gAll)):
    c=patches.Circle((G_Xc[i],G_Yc[i]), G_R[i], fc='b')
    plt.gca().add_patch(c)
    
plt.show()


### 이 방식만으로는 plot이 안된다. 이 method에 대해서는 더 공부해보자.
```
